```python
df = df[(df['Quantity'] > 0)]
df = df[(df['UnitPrice'] > 0)]
```

```python
# Total transaksi per item
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
hasil_agregasi = df.groupby("Description").agg(
    keterlibatan=("InvoiceNo","nunique"),
    jumlah_riil=("Quantity","sum")
).reset_index().sort_values(by="keterlibatan",ascending=False)

# urutkan dari keterlibatan tertinggi
hasil_agregasi = hasil_agregasi.sort_values(by="keterlibatan", ascending=False)

hasil_agregasi
```

```python
median_keterlibatan = hasil_agregasi["keterlibatan"].median()
print("Median keterlibatan:", median_keterlibatan)
```

```python
agregasi = hasil_agregasi[hasil_agregasi['keterlibatan'] > 2]
```

```python
agregasi
```

```python
df = df.merge(agregasi[['Description', 'keterlibatan', 'jumlah_riil']], on='Description', how='left')
```

```python
df
```

```python
# Tanggal referensi = tanggal terakhir transaksi
ref_date = df['InvoiceDate'].max()

# Hitung nilai RFM
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (ref_date - x.max()).days,   # Recency
    'InvoiceNo': 'nunique',                               # Frequency
    'TotalPrice': 'sum'                                   # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Normalisasi kolom Monetary agar tidak terlalu besar
#rfm['Monetary'] = rfm['Monetary'].round(2)

print(rfm.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
from sklearn.cluster import KMeans

sse = []  # sum of squared errors

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1, 10), sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(rfm_scaled)
    sil = silhouette_score(rfm_scaled, labels)
    sil_scores.append(sil)
    print("k =", k, " | Silhouette =", sil)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'Count'}).round(2)

print(cluster_summary)
```

```python
# subset per cluster
cluster_0 = df[df['Cluster'] == 0]
cluster_1 = df[df['Cluster'] == 1]
cluster_2 = df[df['Cluster'] == 2]
```

```python
# mengelompokkan item per transaksi
transactions_cluster0 = cluster_0.groupby('InvoiceNo')['Description'].apply(list)
transactions_cluster1 = cluster_1.groupby('InvoiceNo')['Description'].apply(list)
transactions_cluster2 = cluster_2.groupby('InvoiceNo')['Description'].apply(list)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

# cluster 0
te = TransactionEncoder()
te_ary = te.fit(transactions_cluster0).transform(transactions_cluster0)
df_cluster0_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# FP-Growth cluster 0
frequent_items_0 = fpgrowth(df_cluster0_encoded, min_support=0.07)
rules_0 = association_rules(frequent_items_0, metric="lift", min_threshold=1)
```

```python
rules_0.shape
```

```python
rules_0 = rules_0.sort_values(by='lift', ascending=False).head(10)
rules_0
```

```python
product_map = hasil_agregasi['Description'].to_dict()

def map_product_names(itemset):
    return [product_map.get(i, str(i)) for i in itemset]

rules_0['antecedents'] = rules_0['antecedents'].apply(map_product_names)
rules_0['consequents'] = rules_0['consequents'].apply(map_product_names)

rules_0[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```
